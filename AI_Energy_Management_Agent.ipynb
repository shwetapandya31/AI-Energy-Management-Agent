````python
# ==========================================
# Step 2: Import Required Libraries
# ==========================================

# Library for data analysis
import pandas as pd

# Library for numerical calculations
import numpy as np

# Library for plotting graphs
import matplotlib.pyplot as plt

# Google Gemini AI SDK
from google import genai
from google.genai import types
````

````python
# ==========================================
# Step 3: Configure Google Gemini API
# ==========================================

# Import userdata to securely access API Key
from google.colab import userdata

# Read Gemini API Key
API_KEY = userdata.get("GEMINI_API_KEY")

# Create Gemini Client
client = genai.Client(api_key=API_KEY)
````

````python
# ==========================================
# Step 4: Create Sample Institutional Dataset
# ==========================================

import pandas as pd

# Sample energy consumption data for an educational institution

energy_data = {
    "Location": [
        "Classroom Block",
        "Computer Lab",
        "Library",
        "Administration Office",
        "Cafeteria",
        "Auditorium",
        "Hostel",
        "Science Laboratory"
    ],

    "Energy_Consumption_kWh": [
        180,
        420,
        150,
        110,
        260,
        340,
        510,
        390
    ]
}

# Create DataFrame
df = pd.DataFrame(energy_data)

# Display Dataset
print("Institutional Energy Consumption Dataset")
display(df)
````

````python
# ==========================================
# Step 5: Visualize Energy Consumption
# ==========================================

import matplotlib.pyplot as plt

# Create bar chart
plt.figure(figsize=(10,6))

plt.bar(
    df["Location"],
    df["Energy_Consumption_kWh"],
    color="skyblue"
)

# Chart Title
plt.title("Institutional Energy Consumption")

# X-axis Label
plt.xlabel("Location")

# Y-axis Label
plt.ylabel("Energy Consumption (kWh)")

# Rotate x-axis labels
plt.xticks(rotation=45)

# Show grid
plt.grid(axis="y", linestyle="--", alpha=0.6)

# Display chart
plt.show()
````

````python
# ==========================================
# Step 6: AI Energy Consumption Analysis
# ==========================================

# Convert dataset into text format
energy_text = df.to_string(index=False)

# Prompt for Gemini AI
prompt = f"""
You are an AI Energy Optimization Assistant.

Analyze the following institutional energy consumption dataset.

Dataset:

{energy_text}

Please provide:

1. A summary of the institution's energy usage.
2. Which locations consume the most electricity?
3. Any inefficient energy usage patterns.
4. Three practical recommendations to reduce electricity consumption.
5. Explain how these recommendations support SDG 13 (Climate Action).

Present the response in a professional report format.
"""

# Generate AI response
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

# Display AI-generated report
print(response.text)
````

````python
# ==========================================
# Step 7: Detect High Energy Consumption Areas
# ==========================================

# Define threshold for high energy consumption
threshold = 300

# Filter locations with high energy usage
high_energy = df[df["Energy_Consumption_kWh"] > threshold]

# Display results
print("High Energy Consumption Areas\n")
display(high_energy)
````

````python
# ==========================================
# Step 8: Carbon Emission Estimation
# ==========================================

# Carbon emission factor
# (Approximate value: 0.82 kg CO₂ per kWh)

emission_factor = 0.82

# Calculate carbon emissions
df["Carbon_Emission_kgCO2"] = df["Energy_Consumption_kWh"] * emission_factor

# Calculate total carbon emission
total_emission = df["Carbon_Emission_kgCO2"].sum()

# Display updated dataset
print("Institutional Carbon Emission Report\n")
display(df)

print(f"\nTotal Estimated Carbon Emission: {total_emission:.2f} kg CO₂")
````

````python
# ==========================================
# Step 9: Visualize Carbon Emissions
# ==========================================

import matplotlib.pyplot as plt

# Create figure
plt.figure(figsize=(10,6))

# Plot carbon emissions
plt.bar(
    df["Location"],
    df["Carbon_Emission_kgCO2"],
    color="green"
)

# Chart title
plt.title("Estimated Carbon Emissions by Location")

# Axis labels
plt.xlabel("Location")
plt.ylabel("Carbon Emission (kg CO₂)")

# Rotate x-axis labels
plt.xticks(rotation=45)

# Display grid
plt.grid(axis="y", linestyle="--", alpha=0.6)

# Show chart
plt.show()
````

````python
# ==========================================
# Step 10: Real-Time Energy Monitoring Simulation
# ==========================================

import random
import time

print("Real-Time Institutional Energy Monitoring\n")

# Simulate 10 live readings
for i in range(10):

    location = random.choice(df["Location"])

    energy = random.randint(100, 550)

    print(f"Location: {location}")
    print(f"Current Energy Usage: {energy} kWh")

    print("-" * 45)

    time.sleep(1)
````

````python
# ==========================================
# Step 11: Smart Alerts for Abnormal Energy Usage
# ==========================================

# Define alert threshold
alert_threshold = 400

print("Smart Energy Alerts\n")

# Check every location
for index, row in df.iterrows():

    if row["Energy_Consumption_kWh"] > alert_threshold:

        print(f"⚠ ALERT!")
        print(f"Location : {row['Location']}")
        print(f"Energy Usage : {row['Energy_Consumption_kWh']} kWh")
        print("Recommendation : Immediate energy optimization required.\n")

    else:

        print(f"✅ {row['Location']} - Normal Energy Consumption")
````

````python
# ==========================================
# Step 12: Generate AI Institutional Energy Management Report
# ==========================================

# Convert dataset into text format
energy_text = df.to_string(index=False)

# Prompt for Gemini AI
prompt = f"""
You are an AI Energy Management Assistant.

Analyze the following institutional energy consumption data.

{energy_text}

Generate a professional report with the following sections:

1. Executive Summary
2. Total Energy Consumption
3. Highest Energy Consuming Locations
4. Lowest Energy Consuming Locations
5. Energy Consumption Analysis
6. High Energy Consumption Areas
7. Carbon Emission Summary
8. Energy Saving Recommendations
9. Smart Monitoring Suggestions
10. Sustainability Suggestions
11. SDG 13 (Climate Action) Impact
12. Conclusion

Use proper headings, bullet points and professional formatting.
"""

# Generate AI Report
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

# Store the generated report
final_report = response.text

# Display the AI-generated report
print(final_report)
````

````python
# ==========================================
# Step 13: Save AI Report
# ==========================================

# Save the AI-generated report
with open("Institutional_Energy_Report.txt", "w") as file:
    file.write(final_report)

print("AI Report saved successfully!")
````
